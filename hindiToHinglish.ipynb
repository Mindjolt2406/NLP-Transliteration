```python
def getDictionary():

    Vowels = [a,aa,i,ee,ou,oou,ae,aye,o,ao,aou,an,aan]
    Vowels_ = ["a","aa","i","ee","ou","oou","ae","aye","o","ao","aou","an","aan"]
    Consonants = [ka,kha,ga,gha,rda,cha,chha,ja,jha,iyan,tta,ttha,dda,ta,tha,da,dha,na,pa,fa,ba,bha,ma,ya,ra,la,va,sha,sha2,sa,ha,za,tra]
    Consonants_ = ["ka","kha","ga","gha","rda","cha","chha","ja","jha","iyan","tta","ttha","dda","ta","tha","da","dha","na","pa","fa","ba","bha","ma","ya","ra","la","va","sha","sha","sa","ha","za","tra"]
    Single_matras = [s_an,s_aan,s_aa,s_i,s_ee,s_o,s_ao,s_u,s_oo,s_ae,s_ai,s_aaw,s_aw]
    Single_matras_ = ["an","aan","aa","i","ee","o","ao","u","oo","ae","ai","aaw","aw"]

    vowel_dict = {}
    consonant_dict = {}
    matra_dict = {}

    for j in range(len(Vowels)):
        vowel_dict[Vowels[j]] = Vowels_[j]

    for j in range(len(Consonants)):
        consonant_dict[Consonants[j]] = Consonants_[j]

    for j in range(len(Single_matras)):
        matra_dict[Single_matras[j]] = Single_matras_[j]
    
    return [vowel_dict, consonant_dict, matra_dict]
```

```python
def convertHindiToHinglish(word):
   
    dictionary = getDictionary()
    vowels = dictionary[0]
    consonants = dictionary[1]
    matras = dictionary[2]

    stack = []
    counter = 0
    output = ''

    if ord(word[0]) < 2300:
        return word
    if word[-1] not in vowels and word[-1] not in consonants and word[-1] not in matras and word[-1] != half:
        word = word[:-1]

    # single characters
    if len(word) == 1:
        if word in consonants:
            output = consonants[word]     
            return output[:-1]
        elif word in vowels:
            return vowels[word]

    while counter < len(word):
        if len(stack) == 0:
            stack.append(word[counter])
            counter = counter + 1

            if counter == len(word): # if we have reached the end of the word
                letter = stack.pop()
                if letter in consonants:
                    output = output +  consonants[letter]
                    output = output[:-1]
                if letter in vowels:
                    output = output + vowels[letter]
            continue

        
        letter = word[counter]

        
        if letter == half: # this must always occur after a consonant so the other cases have been ignored
            var = stack.pop()
            output = output + consonants[var]
            output = output[:-1]
            
        elif letter in matras: # only taking care of the matras of the right side
            var = stack.pop()
            output = output + consonants[var]
            output = output[:-1]
            output = output + matras[letter]

        elif letter in consonants or letter in vowels:
            var = stack.pop()
            if var in consonants:
                output = output + consonants[var]
            elif var in vowels:
                output = output + vowels[var]

            if counter != len(word)-1:
                stack.append(letter)
            else:
                if letter in consonants:
                    output = output +  consonants[letter]
                    output = output[:-1]
                
        counter = counter + 1

    return output
```
